```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
from scipy.stats import pareto
from scipy.stats import gaussian_kde
```

```python
plt.rcParams["figure.figsize"] = (15,10)
```

# Problem 7

Simulate $n$ samples from a given distribution $F$, and then use kernel density estimation to estimate the density.

**(a)** $F$ is normal $N(0, \sigma^2)$. Try a few different value for $\sigma^2$.

**(b)** $F$ is Pareto distributed as above. Try a few different values for parameters.

# Solution

## (a)

```python
xs = np.linspace(-5, 5, 1000)

fig, ax = plt.subplots(2, 2, sharex=True, sharey=True)

def KDEplots(N, loc, scale, subplot_axis_arg_a, subplot_axis_arg_b, color):
    
    # Define axis of the subplot
    axis = ax[subplot_axis_arg_a, subplot_axis_arg_b]
    
    # Generate samples
    x = np.random.normal(loc=loc,scale=scale,size=N)
    
    # Plot density F
    axis.fill(xs, density, fc=color, alpha=0.2, label="Distribution F")
    
    # Kernel density estimation
    kde = gaussian_kde(x)
    
    dens = kde(xs)
    
    # plot
    axis.plot(xs, dens, color=color, label="Gaussian kernel")
    axis.plot(x, -0.005 - 0.01 * np.random.random(x.shape[0]), "+k", label="Generated points")
    axis.set_title(f'N = {N} points, std = {scale}')
    axis.legend(loc="upper left")

loc=0
scale=1
color="blue"

density = norm.pdf(xs, loc=loc, scale=scale)

KDEplots(50, loc, scale, 0, 0, color)
KDEplots(500, loc, scale, 0, 1, color)
KDEplots(5000, loc, scale, 1, 0, color)
KDEplots(50000, loc, scale, 1, 1, color)

plt.show()
```

```python
loc=0
scale=0.1
color="orange"

xs = np.linspace(-0.5, 0.5, 1000)

density = norm.pdf(xs, loc=loc, scale=scale)
fig, ax = plt.subplots(2, 2, sharex=True, sharey=True)

KDEplots(50, loc, scale, 0, 0, color)
KDEplots(500, loc, scale, 0, 1, color)
KDEplots(5000, loc, scale, 1, 0, color)
KDEplots(50000, loc, scale, 1, 1, color)

plt.show()
```

## (b)

```python
xs = np.linspace(0, 5, 1000)

fig, ax = plt.subplots(2, 2, sharex=False, sharey=False)

def KDEplots2(N, b, loc, scale, subplot_axis_arg_a, subplot_axis_arg_b, color):
    
    # Define axis of the subplot
    axis = ax[subplot_axis_arg_a, subplot_axis_arg_b]
    
    # Generate samples
    x = pareto.rvs(b, size=N, loc=loc, scale=scale)
    
    # Plot density F
    axis.fill(xs, density, fc=color, alpha=0.2, label="Distribution F")
    
    # Kernel density estimation
    kde = gaussian_kde(x)
    
    dens = kde(xs)
    
    # plot
    axis.plot(xs, dens, color=color, label="Gaussian kernel")
    axis.plot(x, -0.005 - 0.01 * np.random.random(x.shape[0]), "+k", label="Generated points")
    axis.set_title(f'N = {N} points, std = {scale}, b = {b}')
    axis.legend(loc="upper left")

loc=0
scale=1
color="blue"

b = 5

density = pareto.pdf(xs, b, loc=loc, scale=scale)


KDEplots2(50, b, loc, scale, 0, 0, color)
KDEplots2(500, b, loc, scale, 0, 1, color)
KDEplots2(5000, b, loc, scale, 1, 0, color)
KDEplots2(50000, b, loc, scale, 1, 1, color)

plt.show()
```

```python
loc=0
scale=2
color="orange"

b = 5

density = pareto.pdf(xs, b, loc=loc, scale=scale)

fig, ax = plt.subplots(2, 2, sharex=True, sharey=True)


KDEplots2(50, b, loc, scale, 0, 0, color)
KDEplots2(500, b, loc, scale, 0, 1, color)
KDEplots2(5000, b, loc, scale, 1, 0, color)
KDEplots2(50000, b, loc, scale, 1, 1, color)

plt.show()
```

```python
xs = np.linspace(0.75, 1.75, 1000)

loc=0
scale=1
color="red"

b = 25

density = pareto.pdf(xs, b, loc=loc, scale=scale)

fig, ax = plt.subplots(2, 2, sharex=True, sharey=True)


KDEplots2(50, b, loc, scale, 0, 0, color)
KDEplots2(500, b, loc, scale, 0, 1, color)
KDEplots2(5000, b, loc, scale, 1, 0, color)
KDEplots2(50000, b, loc, scale, 1, 1, color)

plt.show()
```
